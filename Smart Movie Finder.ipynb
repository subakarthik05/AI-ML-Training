```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv(r"C:\Users\muthu\Downloads\Movie-dataset\tmdb_5000_credits.csv")
movies_df = pd.read_csv(r"C:\Users\muthu\Downloads\Movie-dataset\tmdb_5000_movies.csv")
```

```python
credits.head()
movies_df.head()
```

```python
print("Credits:",credits.shape)
print("Movies Dataframe:",movies_df.shape)
```

```python
credits_column_renamed = credits.rename(index=str,
columns={"movie_id": "id"})
movies_df_merge = movies_df.merge(credits_column_renamed, on='id')
movies_df_merge.head()
```

```python
movies_cleaned_df = movies_df_merge.drop(columns=['homepage', 'title_x',
'title_y', 'status','production_countries'])
movies_cleaned_df.head()
movies_cleaned_df.info()
```

```python
# Calculate all the components based on the above
v=movies_cleaned_df['vote_count']
R=movies_cleaned_df['vote_average']
C=movies_cleaned_df['vote_average'].mean()
m=movies_cleaned_df['vote_count'].quantile(0.70)
```

```python
movies_cleaned_df['weighted_average']=((R*v)+
(C*m))/(v+m)
movies_cleaned_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create weighted average column
movies_df['weighted_average'] = (movies_df['vote_average'] * movies_df['vote_count']) / (movies_df['vote_count'] + 1000)
```

```python
# Sort movies by weighted average
weight_average = movies_df.sort_values('weighted_average', ascending=False)
```

```python
# Plot the top 10 movies
plt.figure(figsize=(12,6))
axis1 = sns.barplot(
    x=weight_average['weighted_average'].head(10),
    y=weight_average['original_title'].head(10),
    data=weight_average.head(10)
)
plt.xlim(4, 10)
plt.title('Best Movies by Average Votes', weight='bold')
plt.xlabel('Weighted Average Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.tight_layout()
plt.savefig('best_movies.png')
plt.show()
```

```python
# Sort movies by popularity
popularity = movies_df.sort_values('popularity', ascending=False)

# Show top 5
popularity.head()
```

```python

# Sort by popularity using the correct DataFrame
popularity = movies_df.sort_values('popularity', ascending=False)

# Plotting the top 10 most popular movies
plt.figure(figsize=(12,6))
ax = sns.barplot(
    x=popularity['popularity'].head(10),
    y=popularity['original_title'].head(10),
    data=popularity.head(10)
)
plt.title('Most Popular Movies by Votes', weight='bold')
plt.xlabel('Score of Popularity', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.tight_layout()
plt.savefig('best_popular_movies.png')
plt.show()

```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python

scaling = MinMaxScaler()
movie_scaled_df = scaling.fit_transform(
    movies_cleaned_df[['weighted_average', 'popularity']]
)
movie_normalized_df = pd.DataFrame(
    movie_scaled_df,
    columns=['weighted_average', 'popularity']
)
movie_normalized_df.head()
```

```python
movies_cleaned_df[['normalized_weight_average', 'normalized_popularity']] = movie_normalized_df
```

```python
movies_cleaned_df.head()
```

```python
# Calculate hybrid score with 50% weight each
movies_cleaned_df['score'] = (
    movies_cleaned_df['normalized_weight_average'] * 0.5 +
    movies_cleaned_df['normalized_popularity'] * 0.5
)

# Sort the DataFrame by the new score
movies_scored_df = movies_cleaned_df.sort_values('score', ascending=False)

# Display top 20 movies by hybrid score
movies_scored_df[['original_title', 'normalized_weight_average', 'normalized_popularity', 'score']].head(20)
```

```python
scored_df = movies_cleaned_df.sort_values('score',
ascending=False)
```

```python
plt.figure(figsize=(16,6))
```

```python
ax = sns.barplot(x=scored_df['score'].head(10),
y=scored_df['original_title'].head(10), data=scored_df,
palette='deep')
```

```python
# Plot top 10 scored movies
plt.figure(figsize=(12,6))
sns.barplot(
    x=movies_scored_df['score'].head(10),
    y=movies_scored_df['original_title'].head(10),
    data=movies_scored_df.head(10),
    palette='viridis'
)

# Optional axis limit
# plt.xlim(3.55, 5.25)

plt.title('Best Rated & Most Popular Blend', weight='bold')
plt.xlabel('Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.tight_layout()
plt.savefig('hybrid_top_movies.png')
plt.show()
```
